```python
import numpy as np 
X_train = np.load("data/X_train.npy")
X_test = np.load("data/X_test.npy")
X_train.shape, X_test.shape
```

```python
from lrmc.factor_model import GraphConvRNN

model = GraphConvRNN(rank=10, n_iter=5, k_neighbors=2, n_channels=16)
model.fit(X_train)
```

```python
from sklearn.metrics import mean_squared_error

train_mask = X_train > 0
test_mask = X_test > 0

X_train_hat = np.round(model.M) * train_mask
X_test_hat = np.round(model.M) * test_mask

score_train = mean_squared_error(X_train, X_train_hat)
score_test = mean_squared_error(X_test, X_test_hat)
print(score_train, score_test)
```
