```python
import numpy as np
import matplotlib.pyplot as plt

# Create time axis
t = np.linspace(0, 50, 500)   

# Linear trend
trend = 0.5 * t

# Oscillating component
oscillation = 2 * np.sin(2 * np.pi * t / 10)  # period = 10

# Combine them
time_series = trend + oscillation

# Plot
plt.figure()
plt.plot(t, time_series, label="Time Series")
plt.plot(t, trend, '--', label="Trend")
plt.plot(t, oscillation, ':', label="Oscillation")

plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.title("Time Series with Linear Trend + Oscillation")
plt.show()
```
